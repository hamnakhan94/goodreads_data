```python
df = pd.read_csv("./Desktop/goodreads_data.csv", header = 1, parse_dates = [14])
df = df.drop(columns=["ISBN", "ISBN13", "Bookshelves", "Bookshelves with positions", "My Review", "Spoiler", "Private Notes", "Owned Copies"])
df = df.drop(columns=["Book Id", "Author l-f", "Additional Authors", "Binding", "Original Publication Year", "Date Added", "Exclusive Shelf"])
df = df.dropna(subset="Date Read").reset_index(drop=True)
df["Date Read"]=df["Date Read"].dt.year.astype(int)


```

```python
df.to_excel("./Desktop/GoodReads_Clean.xlsx")
```
